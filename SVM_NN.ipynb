```python
# Normalizing data and filtering for minimum minutes (oof this is ugly)
def normalize(data):
    minTime = 10.0
    low_minutes = []
    for i in range(len(data)):
        if data[i][3] < minTime:
            low_minutes.append(i)
    ndata = np.ndarray( ( len(data)-len(low_minutes), len(data[0])))

    j=0
    for i in range(len(data)):
        if i not in low_minutes:
            ndata[j] = data[i]
            j = j+1

    for i in range(len(ndata[0])-1):
        up = max(ndata[:,i])
        lo = min(ndata[:,i])
        ndata[:,i] = (ndata[:,i]-lo)/(up-lo) - 0.5
    return ndata
ndata = normalize(data)
```

```python
# Building MVP distance matrix
def dist(a, b):
    s = 0
    for i in range(len(a)):
        s = s + (a[i] - b[i])**2
    return np.sqrt(s)

l = len(ndata_MVP)
dist_MVP = np.ndarray((l,l))
for i in range(l):
    for j in range(l):
        dist_MVP[i][j] = dist(ndata_MVP[i][:-1], ndata_MVP[j][:-1])
```

```python
# Finding nearest neighbors (ct is how many neighbors)
ct = 5
closest = np.zeros((l,ct)) + 10
close_dist = np.zeros((l,ct)) + 10
for i in range(l):
    for j in range(l):
        if i != j:
            if dist_MVP[i][j] < np.max(close_dist[i]):
                closest[i][np.argmax(close_dist[i])] = j
                close_dist[i][np.argmax(close_dist[i])] = dist_MVP[i][j]
```

```python
# Code for generating printable list of closest MVPs
for i in range(l):
    mID = int(ndata_MVP[i][-1])
    year = raw_data[mID][0]
    name = Stats[mID][2]
    close = []
    for j in range(ct):
        close.append((Stats[int(ndata_MVP[int(closest[i][j])][-1])][2],
                      int(Stats[int(ndata_MVP[int(closest[i][j])][-1])][1])))
    #print(close)
```

```python
# Generate N synthetic MVPs
N = 500
def MVP_gen(N):
    sdata_MVP = np.ndarray((N,len(ndata_MVP[0])))
    c1 = np.random.randint(0,l,N)
    c2 = np.random.randint(0,ct,N)
    for i in range(N):
        data1 = ndata_MVP[c1[i]]
        data2 = ndata_MVP[int(closest[c1[i],c2[i]])]
        sdata_MVP[i][:] = np.random.rand()*(data1-data2) + data2
    return sdata_MVP
sdata_MVP = MVP_gen(N)
```

```python
import torch
import torch.nn as nn
import numpy
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
kfold = KFold(n_splits = 5, shuffle = True, random_state = 0)
```

```python
def split_by_year(n):
    year_groups_ = year_groups[24:]
    data_year_ = data_year[24:]
    year_ct = len(year_groups_)
    year_list = np.random.choice(year_ct,size=int(n),replace=0)
    
    testing_index = []
    for i in range(n):
        grp = year_groups_[year_list[i]]
        for j in range(len(grp)):
            testing_index.append(grp[j])
            
    other_index = np.setdiff1d(range(len(cdata)),testing_index)        
    
    testing_data = np.array(cdata[testing_index])
    testing_labels = np.array(clabels[testing_index])
    other_data = np.array(cdata[other_index])
    other_labels = np.array(clabels[other_index])
    return testing_data, testing_labels, other_data, other_labels, year_list
```

```python
# Find the "true score" of our logistic regression by finding the player with
# the highest likelihood to be MVP each year
def true_score(ry_pred):
  ccorrect = []
  counter = 0
  for i in year_list:
    grp = year_groups[i]
    if len(grp) > 1:
        counter += 1
        argm = np.argmax(ry_pred[grp])
        mvp_choice = grp[argm]
        if int(ndata[mvp_choice][-1]) in MVP_index:
            ccorrect.append(i)
  true_score1 = len(ccorrect)/counter
  print('True Score for test:',np.round(true_score1,5))
  ccorrect = []
  counter = 0
  for i in range(len(year_groups)):
    grp = year_groups[i]
    if len(grp) > 1:
      counter += 1
      argm = np.argmax(ry_pred[grp])
      mvp_choice = grp[argm]
      if int(ndata[mvp_choice][-1]) in MVP_index:
        ccorrect.append(i)
  true_score2 = len(ccorrect)/counter
  print('True Score for all:',np.round(true_score2,5))
  return true_score1, true_score2
```

```python
def predict(models, test_data):
  rx = torch.FloatTensor(ndata[:,:-1])
  cx = torch.FloatTensor(test_data[:,:-1])
  ry_pred = numpy.zeros((11801,1))
  cy_pred = numpy.zeros((len(test_data), 1))
  for i in range(5):
    ry_pred = ry_pred + models[i](rx).detach().numpy()
    cy_pred = cy_pred + models[i](cx).detach().numpy()
  ry_pred = ry_pred / 5
  cy_pred = cy_pred / 5
  return ry_pred, cy_pred
  
```

```python
def plot(loss_lists_train, loss_lists_test):
  plt.subplot(211)
  plt.xlabel('Number of iterations')
  plt.ylabel('Training loss')
  for i in range(len(loss_lists_train)):
    plt.plot(loss_lists_train[i])
  plt.subplot(212)
  plt.xlabel('Number of iterations')
  plt.ylabel('Test loss')
  for i in range(len(loss_lists_test)):
    plt.plot(loss_lists_test[i])
  plt.show()
```

```python
def evaluate(rpred, cpred):  
  my_cpos = np.where(cpred>=0.5)[0]
  my_cMVP = np.where(test_labels==1)[0]

  my_rpos = np.where(rpred>=0.5)[0]
  my_rMVP = np.where(labels==1)[0]

  my_ccorrect = np.intersect1d(my_cpos,my_cMVP)
  my_rcorrect = np.intersect1d(my_rpos,my_rMVP)
  my_rfpos = np.setdiff1d(my_rpos,my_rcorrect)
  return len(my_ccorrect), len(my_rcorrect), len(my_rfpos)
```

```python
# *** Note: In order to enable GPU in Google Colab, you need to click "edit", then click "Notebook settings", and enable "Hardware accerlerator" first ***
def train(train_data, train_labels):
  n_in, n_h, n_out, batch_size = len(cdata[0])-1, 16, 1, 128
  device = torch.device("cuda" if torch.cuda.is_available() else "cpu") # enable GPU
  models = []
  for i in range(5):
    models.append(nn.Sequential(nn.Linear(n_in, n_h),
                                nn.LeakyReLU(),
                                nn.Linear(n_h, n_h),
                                nn.LeakyReLU(),
                                nn.Linear(n_h, n_h),
                                nn.LeakyReLU(),
                                nn.Linear(n_h, n_out),
                                nn.Sigmoid()))
  loss_lists_train = []
  loss_lists_test = []
  count = 0
  for train, test in kfold.split(train_data[:,:-1]):
    loss_list_train = []
    loss_list_test = []
    x = torch.FloatTensor(train_data[train,:-1])
    x_test = torch.FloatTensor(train_data[test,:-1])
    y = torch.FloatTensor(train_labels[train])
    y_test = torch.FloatTensor(train_labels[test])
    training_model = models[count].to(device)
    optimizer = torch.optim.SGD(training_model.parameters(), lr=0.1, momentum=0.9)
    for epoch in range(5000):
      # Zero the gradients
      optimizer.zero_grad()
      x = x.to(device) # feed input to GPU
      y = y.to(device) # feed labels to GPU
      x_test = x_test.to(device)
      y_test = y_test.to(device)
      # Forward Propagation
      y_pred = training_model(x)
      y_pred_test = training_model(x_test)
      # Compute and print loss
      loss = nn.functional.binary_cross_entropy(y_pred, y)
      loss_test = nn.functional.binary_cross_entropy(y_pred_test, y_test)
      loss_list_train.append(loss.item())
      loss_list_test.append(loss_test.item())
      # perform a backward pass (backpropagation)
      loss.backward()
      # Update the parameters
      optimizer.step()
    models[count] = training_model.cpu()
    loss_lists_train.append(loss_list_train)
    loss_lists_test.append(loss_list_test)
    count = count + 1
  plot(loss_lists_train, loss_lists_test)
  return models
```

```python
true_score_test_list = []
true_score_all_list = []
rcorrect_list = []
fpos_list = []
for i in range(10):
  print("Iteration: ", i)
  models = train(train_data, train_labels)
  r_pred, c_pred = predict(models, test_data)
  ccorrect, rcorrect, fpos = evaluate(r_pred, c_pred)
  true_score_test, true_score_all = true_score(r_pred)  
  true_score_test_list.append(true_score_test)
  true_score_all_list.append(true_score_all)
  rcorrect_list.append(rcorrect)
  fpos_list.append(fpos)
```
